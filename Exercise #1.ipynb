```python
import matplotlib.pyplot as plt
import numpy as np

year = [1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
GDP = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
plt.plot(year, GDP, 'ro', linestyle='dashed')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.title('Real US GDP (in trillions)')
plt.show()
```

A possible mathematical relationship between GDP and year is a quadratic relationship. At small values of x, the values of y are relatively close together but as we get to higher values of x the values of y are further apart from eachother. In other words, the slope becomes steeper towards higher values of x. The quadratic relation with the form, $$y=a^2x+bx+c,$$ fits this relationship way better than a linear relationship because it has a parabola shape.

```python
z = np.polyfit(year, GDP, 2)
p = np.poly1d(z)
plt.plot(year, GDP, 'ro', year, p(year), '-')
plt.show()
```

The blue line represents a perfectly quadratic relationship between x and y and the red dots represent the actual values of x and y.
